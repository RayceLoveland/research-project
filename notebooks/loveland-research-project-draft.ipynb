&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Throughout this writing, it has been established that the German war production was entirely on the wrong footing, but there are reasons for that.  Overy said, Hitler could have had twice the weapons in mid-1941 than he had, if the converted resources had been more productively exploited.(Overy 1988, 630) John Parshalls lecture on German tank production is a good representation of the broader outlook of their overall production. Manufacturers had to work with the German army/ infantry to fit their specifications, which demanded a very high-quality tank. Craftsmen were brought to work on these tanks. Tanks were built as stationary assemblies, not assembly lines, which are usually used for luxurious products. Factories were not modular places, like the Soviets, who could convert tractor factories into tank factories if necessary. Tank plan designs would be rushed to then be a pain to produce, had to use chalk to draw on tank to remind what step they were in the process, had to make on average 2-3 modifications a week (meant a tank six rows down would be slightly different), and had to add personal touches requested like rain cover for the commander is one example. Production was aimed at making the perfect vehicle. One stat that Parshall showed was the split effectiveness of each output: one M1 Sherman cost around 10,000 man-hours and 33,500 dollars, while a Tiger cost about 300,000 man-hours and around 320,000 dollars (Parshall 2013). A cost value that isnt ultimately wasnt worth it.
