<u>Fig 1.

```python
# Get top 10 countries by people mobilized
top10 = people_mobilized.sort_values(by="Value", ascending=False).head(10)
# Assign 'red' for Axis and 'blue' for Allied directly
color_map = {"Allied": "blue", "Axis": "red"}
bar_colors = top10["Category"].map(color_map)
# Create bar chart with manual coloring
fig = px.bar(
    top10,
    x="Country",
    y="Value",
    title="Top 10 Countries by Number of People Mobilized in WWII",
    labels={"Value": "People Mobilized", "Country": "Country"},
    text="Value"
)
fig.update_traces(
    marker_color=bar_colors,
    texttemplate='%{text:,}',
    textposition='outside'
)
fig.update_layout(
    yaxis_title="Mobilized Personnel",
    xaxis_title="Country",
    showlegend=False,  
    margin=dict(t=100),
    yaxis=dict(range=[0, top10["Value"].max() * 1.2]),
    uniformtext_minsize=8,
    uniformtext_mode='hide'
)
fig.show()
```

<u>Fig 2.

```python
# Define the population data
population_data = pd.DataFrame({
    "Country": [
        "Germany and Austria",
        "Germany's European allies",
        "German-occupied countries"
    ],
    "Value": [76000000, 78000000, 129000000]
})
# Create the pie chart
fig = px.pie(
    population_data,
    names="Country",
    values="Value",
    title="German Wartime Population Distribution (1941)",
    color="Country",
    color_discrete_map={
        "Germany and Austria": "red",       
        "Germany's European allies": "orange",  
        "German-occupied countries": "blue"   
    }
)
fig.update_traces(textposition='inside', textinfo='label+percent')
fig.update_layout(
    showlegend=True
)
fig.show()
```

*Fig 3*

```python
# Define data and multiply to convert from 'thousands' to actual tool counts
metal_tools_data = pd.DataFrame({
    "Country": ["United States", "Germany", "Soviet Union", "Great Britain"],
    "Value": [112, 125, 58.4, 0]
})
metal_tools_data["Value"] *= 1000  # Convert to actual machine counts
# Create bar chart
fig = px.bar(
    metal_tools_data,
    x="Country",
    y="Value",
    title="WWII Metal-Working Tool Production in 1940",
    labels={"Value": "Machine Tools Produced", "Country": "Country"},
    text="Value"
)
# Show value labels with commas
fig.update_traces(
    texttemplate='%{text:,}',
    textposition='outside'
)
# Fix Y-axis and margins for visibility
fig.update_layout(
    yaxis_title="Number of Machine Tools",
    xaxis_title="Country",
    showlegend=False,
    margin=dict(t=100),
    yaxis=dict(range=[0, metal_tools_data["Value"].max() * 1.2])
)
fig.show()
```

*Fig 4*

```python
# Filter data between 1933 and 1945
filtered_steel = crude_steel_production[crude_steel_production["Year"].between(1933, 1945)]
# Create grouped bar chart 
fig = px.bar(
    filtered_steel,
    x="Country",
    y="Value",
    color="Year",
    title="Crude Steel Production by Country (1933–1945)",
    labels={"Value": "Crude Steel Produced (tons)", "Country": "Country", "Year": "Year"},
    barmode="group"
)
# Style the layout to reduce crowding
fig.update_layout(
    yaxis_title="Tons of Crude Steel",
    xaxis_title="Country",
    bargap=0.25,        
    bargroupgap=0.15,   
    width=1000,
    height=500,
    margin=dict(t=100),
    showlegend=True     
)
# Remove text labels from bars
fig.update_traces(text=None)
fig.show()
```

*Fig 5*

```python
# Filter for specific countries and years
countries = ["Germany", "Luxembourg", "France", "United Kingdom"]
filtered_ore = iron_ore_production[
    (iron_ore_production["Country"].isin(countries)) &
    (iron_ore_production["Year"].between(1933, 1945))
]
# Create line chart
fig = px.line(
    filtered_ore,
    x="Year",
    y="Value",
    color="Country",
    title="Iron Ore Production (1933–1945)",
    labels={"Value": "Iron Ore Produced (tons)", "Country": "Country", "Year": "Year"}
)
# Stretch and style
fig.update_layout(
    width=900,
    height=500,
    margin=dict(t=80),
    yaxis_title="Tons of Iron Ore",
    xaxis_title="Year",
    showlegend=True
)
fig.show()
```

```python
# Filter years 1939 to 1944 
filtered_spending = military_spending[military_spending["Year"].between(1939, 1944)]
# Create the line chart
fig = px.line(
    filtered_spending,
    x="Year",
    y="Value",
    color="Country",
    title="Military Spending as Share of National Income (1939–1944)",
    labels={"Value": "Spending", "Year": "Year", "Country": "Country"}
)
# Format Y-axis as percent and set layout
fig.update_layout(
    yaxis_tickformat=".0%", 
    yaxis_title="Percent of National Income",
    xaxis_title="Year",
    width=900,
    height=500,
    margin=dict(t=80),
    showlegend=True
)

fig.show()
```

```python
# Get top 4 countries by total naval vessel output
top4 = (
    naval_production.groupby("Country")["Value"]
    .sum()
    .nlargest(4)
    .index
)

# Filter original dataset for those countries
top4_naval = naval_production[naval_production["Country"].isin(top4)]

# Step 3: Create the grouped bar chart
fig = px.bar(
    top4_naval,
    x="Country",
    y="Value",
    color="Year",
    barmode="group",
    title="Top 4 Countries by Naval Vessel Production (1939–1945)",
    labels={"Value": "Vessels Produced", "Country": "Country", "Year": "Year"}
)

# Layout adjustments
fig.update_layout(
    width=1000,
    height=500,
    margin=dict(t=80),
    bargap=0.25,
    bargroupgap=0.1,
    yaxis_title="Number of Naval Vessels",
    xaxis_title="Country",
    showlegend=True,
    legend_title_text="Year"
)

fig.show()
```

```python
 #Identify top 4 countries by total aircraft output
top4_aircraft_countries = (
    aircraft_production.groupby("Country")["Value"]
    .sum()
    .nlargest(4)
    .index
)
 #Filter dataset to those top 4 countries
top4_aircraft = aircraft_production[aircraft_production["Country"].isin(top4_aircraft_countries)]
#Create grouped bar chart
fig = px.bar(
    top4_aircraft,
    x="Country",
    y="Value",
    color="Year",
    barmode="group",
    title="Top 4 Countries by Aircraft Production (1939–1945)",
    labels={"Value": "Aircraft Produced", "Country": "Country", "Year": "Year"}
)
 #Layout styling
fig.update_layout(
    width=1000,
    height=500,
    margin=dict(t=80),
    bargap=0.25,
    bargroupgap=0.1,
    yaxis_title="Number of Aircraft",
    xaxis_title="Country",
    showlegend=True,
    legend_title_text="Year"
)
fig.show()
```

```python
import plotly.express as px

# Step 1: Find top 4 countries by total tank production
top4_tank_countries = (
    tank_production.groupby("Country")["Value"]
    .sum()
    .nlargest(4)
    .index
)

# Step 2: Filter to include only top 4 countries
top4_tanks = tank_production[tank_production["Country"].isin(top4_tank_countries)]

# Step 3: Create grouped bar chart
fig = px.bar(
    top4_tanks,
    x="Country",
    y="Value",
    color="Year",
    barmode="group",
    title="Top 4 Countries by Tank and Self-Propelled Gun Production (1939–1945)",
    labels={"Value": "Tanks Produced", "Country": "Country", "Year": "Year"}
)
#Layout adjustments
fig.update_layout(
    width=1000,
    height=500,
    margin=dict(t=80),
    bargap=0.25,
    bargroupgap=0.1,
    yaxis_title="Number of Tanks",
    xaxis_title="Country",
    showlegend=True,
    legend_title_text="Year"
)
fig.show()
```
